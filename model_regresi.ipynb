```python

from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
for col in df.columns:
    if df[col].dtype == 'object':
        df[col] = le.fit_transform(df[col])
```

```python
import pandas as pd
import statsmodels.api as sm
from statsmodels.stats.outliers_influence import variance_inflation_factor

X_train_resampled_d = sm.add_constant(df.iloc[:, 2:11])

vif_data = pd.DataFrame()
vif_data["Fitur"] = X_train_resampled_d.columns
vif_data["VIF"] = [variance_inflation_factor(X_train_resampled_d.values, i) for i in range(X_train_resampled_d.shape[1])]

print(vif_data)
```

```python
import pandas as pd
import statsmodels.api as sm
from statsmodels.stats.outliers_influence import variance_inflation_factor

X_train_resampled_d = sm.add_constant(df.iloc[:, 2:11])

vif_data = pd.DataFrame()
vif_data["Fitur"] = X_train_resampled_d.columns
vif_data["VIF"] = [variance_inflation_factor(X_train_resampled_d.values, i) for i in range(X_train_resampled_d.shape[1])]

print(vif_data)
```

```python
from scipy.stats import kstest

for column in df.iloc[:, 2:11].columns:
    stat, p = kstest(df[column], 'norm')
    print(f'Kolmogorov-Smirnov test for {column}:')
    print(f'Statistic: {stat}, p-value: {p}')
    if p < 0.05:
        print(f'Reject null hypothesis: {column} tidak terdistribusi secara normal')
    else:
        print(f'Tidak cukup bukti untuk menolak null hypothesis: {column} terdistribusi secara normal')
    print('\n')
```

```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import PowerTransformer

# Inisialisasi PowerTransformer (pilihan metode: 'yeo-johnson' atau 'box-cox')
pt = PowerTransformer(method='yeo-johnson')

# Fit dan transformasi data
df = pd.DataFrame(pt.fit_transform(df), columns=df.columns)

```

```python
from scipy.stats import kstest

for column in df.iloc[:, 2:11].columns:
    stat, p = kstest(df[column], 'norm')
    print(f'Kolmogorov-Smirnov test for {column}:')
    print(f'Statistic: {stat}, p-value: {p}')
    if p < 0.05:
        print(f'Reject null hypothesis: {column} tidak terdistribusi secara normal')
    else:
        print(f'Tidak cukup bukti untuk menolak null hypothesis: {column} terdistribusi secara normal')
    print('\n')
```

```python
from linearmodels.panel import RandomEffects

# Pastikan data memiliki MultiIndex dengan group dan year
data = df.set_index(['daerah', 'tahun'])

# Regresi panel dengan efek acak (random effects)
model_re = RandomEffects.from_formula('selang_waktu ~ 1 + indeks_pembangunan_manusia + tingkat_pengangguran_terbuka + jumlah_tindak_pidana + jumlah_penduduk_miskin', data=data)
result_re = model_re.fit()

# Tampilkan summary hasil regresi
print(result_re.summary)
```

```python
import pandas as pd
import numpy as np
import statsmodels.api as sm
import statsmodels.formula.api as smf

# Regresi panel dengan efek tetap (fixed effects)
model_fe = smf.ols('selang_waktu ~ 1 + indeks_pembangunan_manusia + tingkat_pengangguran_terbuka + jumlah_tindak_pidana + jumlah_penduduk_miskin + C(daerah) + C(tahun)', data=df).fit()

# Tampilkan summary hasil regresi
print(model_fe.summary())
```
